```python
import numpy as np
import os

data_dir = "/content/drive/MyDrive/ML_Project/Data"
model_dir = data_dir
os.makedirs(model_dir, exist_ok=True)

X = np.load(os.path.join(data_dir, "input.npy"))    # (N,1,44,128)
y = np.load(os.path.join(data_dir, "output.npy"))   # (N,)

train_idx = np.load(os.path.join(data_dir, "train_idx.npy"))
test_idx  = np.load(os.path.join(data_dir, "test_idx.npy"))


def feature_sum(arr):
    #(1,44,128)
    return float(arr.sum())

#train
xtrain = np.zeros(len(train_idx))
ytrain = np.zeros(len(train_idx))



for k in range(len(train_idx)):
    i = int(train_idx[k])
    xtrain[k] = feature_sum(X[i])
    ytrain[k] = float(y[i])

#test features
xtest = np.zeros(len(test_idx))
ytest = np.zeros(len(test_idx))

for k in range(len(test_idx)):
    i = int(test_idx[k])
    xtest[k] = feature_sum(X[i])
    ytest[k] = float(y[i])

# least square train
mean_x = float(xtrain.mean())
mean_y = float(ytrain.mean())

num = 0.0
den = 0.0

for k in range(len(xtrain)):
    dx = xtrain[k] - mean_x
    dy = ytrain[k] - mean_y


    num += dx * dy
    den += dx * dx


a = num / den
b = mean_y - a * mean_x


np.savez(os.path.join(model_dir, "regression_sumX_linear.npz"),
         a=np.array([a], dtype=np.float64),
         b=np.array([b], dtype=np.float64))


print("a =", a)
print("b =", b)

# inference
ypred = np.zeros(len(xtest))
for k in range(len(xtest)):
    ypred[k] = a * xtest[k] + b


np.save(os.path.join(model_dir, "y_true_test_regression.npy"), ytest)
np.save(os.path.join(model_dir, "y_pred_test_regression.npy"), ypred)



mse = float(np.mean((ypred - ytest) ** 2))
rmse = float(np.sqrt(mse))
mae = float(np.mean(np.abs(ypred - ytest)))


print("RMSE:", rmse)
print("MAE :", mae)


```
