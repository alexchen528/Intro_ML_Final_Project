```python
import os
import glob
import numpy as np

CHUNK_DIR = "/ceph/work/SATORI/alex/intro_ML/chunks"
OUT_DIR   = "/ceph/work/SATORI/alex/intro_ML"
os.makedirs(OUT_DIR, exist_ok = True)

X_files = sorted(glob.glob(os.path.join(CHUNK_DIR, "X_chunk_*.npy")))
y_files = sorted(glob.glob(os.path.join(CHUNK_DIR, "y_chunk_*.npy")))


X_list = []
y_list = []

total_events = 0

for x_file, y_file in zip(X_files, y_files):
    X_data = np.load(x_file)
    y_data = np.load(y_file)

    if X_data.shape[0] == 0:
        continue

    X_list.append(X_data)
    y_list.append(y_data)
    total_events += X_data.shape[0]

print("total events", total_events)

X = np.concatenate(X_list, axis=0).astype(np.float32)  # (N, 1, 44, 128)
y = np.concatenate(y_list, axis=0).astype(np.float32)  # (N,)

print(X.shape)
print(y.shape)
```

```python
OUT_DIR = "/dicos_ui_home/alex/intro_ML_project/data"
os.makedirs(OUT_DIR, exist_ok = True)

np.save(os.path.join(OUT_DIR, "input.npy"), X)
np.save(os.path.join(OUT_DIR, "output.npy"), y)

```
